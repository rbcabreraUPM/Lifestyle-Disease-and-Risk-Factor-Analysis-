```python
# needed imports
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
# generate two clusters: a with 100 points, b with 50:
#np.random.seed(4711)  # for repeatability of this tutorial
#a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
#b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
#X = np.concatenate((a, b),)

df = pd.read_csv('src_dest_count.csv')
label = df['src'].tolist()
df = df.drop(['src','des'], axis=1)
np_mat = df.as_matrix()

```

```python
#randomMatrix = np.random.uniform(-10,10,size=(20,3))

linked = linkage(np_mat, 'ward')



plt.figure(figsize=(60, 60))
dendrogram(
            linked,
          #  orientation='right',
            labels=label,
            distance_sort='descending',
            show_leaf_counts=False
          )
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.show()
```

```python
df1 = pd.read_csv('src_dest_count.csv')
```

```python
count_df = df1.groupby('des').count()
count_df = count_df.drop(['des'], axis=1)
```

```python
type(list(df1.src.unique()))
```

```python

np_mat = count_df.as_matrix()
linked = linkage(np_mat, 'ward')



plt.figure(figsize=(8, 8))
dendrogram(
            linked,
            orientation='right',
            labels=list(df1.src.unique()),
            distance_sort='descending',
            show_leaf_counts=False
          )
plt.title('Number of Occurences of a Lifestyle Disease with Common Risk Factor')
plt.xlabel('Lifestyle Disease')
plt.ylabel('Number of Occurences')
plt.show()
```
